```python
#Passo4- Categorizar as variáveis clínicas
#hot enconding (sklearn) OU get.dummies (pandas)

def categorize_clinical_data():
    df= pd.read_csv('Clinical_Data_nii.csv')
    
    cat_columns = ["Sex", "Primary Site", "T-Stage", "N-Stage", "TNM group Stage", "Therapy", "HPV status", "Surgery"]
    
    df_processed = pd.get_dummies(df, columns=cat_columns)
    
    
    df_processed = df_processed[['Patient', 'Sex_F',  'Sex_M', 'Age', 'Primary Site_Hypopharynx', 
                                 'Primary Site_Larynx', 'Primary Site_Nasopharynx', 'Primary Site_Oropharynx', 
                                 'Primary Site_unknown', 'T-Stage_T1', 'T-Stage_T2', 'T-Stage_T3', 'T-Stage_T4', 
                                 'T-Stage_T4a', 'T-Stage_T4b', 'T-Stage_Tx', 'N-Stage_N0', 'N-Stage_N1', 'N-Stage_N2', 'N-Stage_N2a',
                                 'N-Stage_N2b', 'N-Stage_N2c', 'N-Stage_N3', 'N-Stage_N3b',
                                 'TNM group Stage_Stage I', 'TNM group Stage_Stage II', 'TNM group Stage_Stage IIB',
                                 'TNM group Stage_Stage III', 'TNM group Stage_Stage IV', 'TNM group Stage_Stage IVA',
                                 'TNM group Stage_Stage IVB', 'TNM group Stage_unknown', 'HPV status_0', 'HPV status_1',
                                 'HPV status_unknown', 'Lymph Nodes', 'Surgery_0', 'Surgery_1', 'Surgery_unknown',
                                 'Time – diagnosis to CT sim (days)', 'Time – diagnosis to start treatment (days) ', 'Time – treatment (days)', 
                                 'Time – diagnosis to last follow-up(days)', 'Therapy_chemo radiation', 'Therapy_radiation',
                                 'Locoregional', 'Distant','Death']]
    
    
    columns= ['Age', 'Time – treatment (days)', 'Time – diagnosis to last follow-up(days)']
    for column in columns:
        df_processed[column] = (df_processed[column] - df_processed[column].min()) / (df_processed[column].max() - df_processed[column].min())
    

    df_processed.to_csv('Clinical_Data_nii_final.csv', index=False)

    

    
categorize_clinical_data()   
```

```python
#Passo5- Juntar as 3 labels ao 1 ficheiro (Normalized_Features)

def Normalized_Features_Labels():
    df = pd.read_csv('All_Filters_Features_Clean_Normalized.csv')
    
    df1= pd.read_csv('Clinical_Data_nii.csv')
    
    result = pd.concat([df, df1], axis=1)
    
    final= result.drop(columns= ['Patient', 'Sex', 'Age', 'Primary Site', 'T-stage', 'N-stage',
                                 'TNM group stage', 'Time – diagnosis to CT sim (days)',
                                 'Time – diagnosis to start treatment (days) ',
                                 'Time – treatment (days)', 'Therapy',
                                 'Time – diagnosis to last follow-up(days)', 'HPV status', 'Lymph Nodes', 'Surgery'])
    
    final.to_csv('All_Filters_Features_Clean_Normalized_Labels.csv', index=False)
    
    
    return final

Normalized_Features_Labels()
```
